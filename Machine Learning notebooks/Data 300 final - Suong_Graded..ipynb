### RANDOM FOREST

### 1.

```python
%%time
X = bank.drop('Bankrupt', axis=1)  
y = bank['Bankrupt'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

rf_classifier_1 = RandomForestClassifier()

rf_classifier_1.fit(X_train, y_train)

y_pred = rf_classifier_1.predict(X_test)

print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("DO NOT USE ACCURACY HERE TO COMPARE MODELS")
```

### 2. change parameters 

```python
%%time
X = bank.drop('Bankrupt', axis=1)  
y = bank['Bankrupt'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

rf_classifier_2 = RandomForestClassifier(n_estimators=100, max_depth=10)

rf_classifier_2.fit(X_train, y_train)

y_pred = rf_classifier_2.predict(X_test)

print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("DO NOT USE ACCURACY HERE TO COMPARE MODELS")
```

#### Higher prescision, recall and f1-score is better. Generally, the 2 models performed relatively the same. Model 1 is a little better.

### 3. Model 1 takes longer time. (Model 1: 3.59s , model 2: 3.17s)

### 4. I did not change criterion before. To compare 'Gini' and ' Entropy':

```python
param_grid = {'criterion': ['gini', 'entropy']}

grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring='precision')
grid_search.fit(X, y)

print("Best Parameters:", grid_search.best_params_)
```

#### 'entropy' is better

### 5.

```python
tree_rules = export_text(rf_classifier_2.estimators_[0], feature_names=list(X.columns))
print("Decision Rules for One Tree:")
print(tree_rules)
```

### 6. The most important features for classification :

```python
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': rf_classifier_2.feature_importances_})
top_3_features = feature_importance_df.nlargest(3, 'Importance')

print("Top 3 most important features for classification:")
print(top_3_features[['Feature', 'Importance']])
```

# PART 2 

```python
# undersampling
bank_undersampled = pd.concat([resample(bank[bank['Bankrupt'] == 0], n_samples=220, replace=True), bank[bank['Bankrupt'] == 1]])
bank_undersampled['Bankrupt'].value_counts()
```

```python
%%time
X = bank_undersampled.drop('Bankrupt', axis=1)  
y = bank_undersampled['Bankrupt'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

rf_classifier= RandomForestClassifier()

rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
%time
svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)

y_pred = svm_classifier.predict(X_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("DO NOT USE ACCURACY HERE TO COMPARE MODELS")
```
