```python
# Import modules
import pandas as pd # v 1.4.4
import numpy as np # v 1.21.5
import matplotlib.pyplot as plt # v 3.5.2

import scikitplot as skplt #v 0.3.7

from sklearn.model_selection import train_test_split #v 1.0.2
from sklearn.linear_model import LogisticRegression #v 1.0.2
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis #v 1.0.2
from sklearn import metrics  #v 1.0.2

from sklearn.utils import resample #v 1.0.2
from imblearn.over_sampling import SMOTE #v 0.11.0
```

### Load data

```python
loan = pd.read_csv("https://raw.githubusercontent.com/KennedyOdongo/DATA-300-Statistical-Machine-Learning-Fall-2023-/main/Data/loan_default_data_set.csv")
loan
```

### Exploratory Data Analysis (EDA)

## Missing values 

Based on the return, we saw that 
- pct_card_over_50_uti  has 1958 missing values
- rep_income has 1559 missing values
- rep_education has 1 missing value

#### I decided to replace the missing values in the *pct_card_over_50_uti*, *rep_income*, variables with the mean of each of that column,  and *rep_education* with the mode of that column

```python
loan.isna().sum()
```

```python
loan['pct_card_over_50_uti'].fillna(loan['pct_card_over_50_uti'].mean(), inplace=True) #replace NA with the mean 
loan['rep_income'].fillna(loan['rep_income'].mean(), inplace=True) #replace NA with the mean
loan['rep_education'].fillna(loan['rep_education'].mode()[0], inplace=True) #replace NA with the mode
```

```python
#check missing value again
loan.isna().sum()
```

## Classes are imbalance, Def_ind = 1 is minority

```python
loan['Def_ind'].value_counts(normalize = True)
```

```python
loan = pd.get_dummies(loan, columns=['rep_education'], drop_first=True) 
```

# 1. Create 3 samples of the data

### a. First sample created by oversampling the minority class so that the number of observations of the minority and majority class are the same (18,000 for each class) => There are 36, 000 observations in total. 

```python
majority = loan[loan['Def_ind']==0]
minority = loan[loan['Def_ind']==1]

minority_oversampled = resample(minority,n_samples=16000,replace=True)
loan_oversampled = pd.concat([minority_oversampled, minority, majority])

# check distribution
loan_oversampled['Def_ind'].value_counts()
```

### b. Second sample created by undersampling the majority class (2000 observations for each class) => There are have 4000 observations in total. 

```python
majority_undersampled = resample(majority,n_samples=2000,replace=True)
loan_undersampled = pd.concat([majority_undersampled, minority])

# check distribution
loan_undersampled['Def_ind'].value_counts()
```

### c. Third sample created by using SMOTE

```python
X = loan.drop("Def_ind", axis = 1)
y = loan["Def_ind"]

X_resampled, y_resampled = SMOTE(sampling_strategy = 'auto').fit_resample(X, y)

# new data set and check distribution
loan_SMOTE = pd.DataFrame(X_resampled, columns=X.columns)
loan_SMOTE['Def_ind'] = y_resampled

loan_SMOTE['Def_ind'].value_counts()
```

# 2, 3, 4. 
## Logistic regression (LR) and Linear Discriminant Analysis (LDA) 
## Compare model performance on accuracy, recall, and precision
## Plot an AUC_ROC curve 

### define a function that fit the models then plot auc curve, and reutrn scores for each models for one sample

```python
def compare_models(samples, model1, model2):
    '''With 
    samples : samples = [(X1, y1), (X2, y2), (X3, y3)]
    model1,2: LogisticRegression(), LinearDiscriminantAnalysis() for example'''
    
    results = []

    for idx, (X, y) in enumerate(samples, 1):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
        
        for model in [model1, model2]:
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)

            # Plot AUC_ROC
            y_proba = model.predict_proba(X_test)[::,1]
            fpr, tpr, _ = metrics.roc_curve(y_test, y_proba)
            auc = metrics.roc_auc_score(y_test, y_proba)
            plt.plot(fpr, tpr, label=f"Sample{idx} - {model}, AUC={auc:.2f}")
            
            # Accuracy, recall, and precision score
            accuracy = metrics.accuracy_score(y_test, y_pred)
            recall = metrics.recall_score(y_test, y_pred)
            precision = metrics.precision_score(y_test, y_pred)
            
            results.append([f"Sample{idx}", str(model), accuracy, recall, precision])
    
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc=4)
    plt.show()
    
    # return dataframe
    df_result = pd.DataFrame(results, columns=['Sample', 'Model', 'Accuracy', 'Recall', 'Precision'])
    
    return df_result
```

## compare models

```python
X1 = loan_oversampled.drop('Def_ind', axis=1)
y1 = loan_oversampled['Def_ind']

X2 = loan_undersampled.drop('Def_ind', axis=1)
y2 = loan_undersampled['Def_ind']

X3 = loan_SMOTE.drop('Def_ind', axis=1)
y3 = loan_SMOTE['Def_ind']

samples = [(X1, y1), (X2, y2), (X3, y3)]

compare_models(samples, LogisticRegression(), LinearDiscriminantAnalysis()) 
```

## 5. Which is the best model?
 
### based on AUC curve:

The higher the auc, the better the model is. Or in other words, the curve closer to top left is better.

For all samples, LDA has a higher AUC than LR. 

Espically the highest AUC is when fitting LDA for Sample3

### based on accuracy, recall, precision score:

For example of SMOTEsample:

 Accuracy score = 0.77  ~ the model is correct in 77.0 % of its predictions.
 
 Precision score = 0.75 ~  75.0 % of positive predictions are correct.
 
 Recall score = 0.82 ~ The model captures 82.0 %of actual positive cases.
 
 
For all samples, LDA model has higher scores (all 3 scores)

Espically sample3 (using SMOTE to creat this dataset) has the highest score for all 3 scores.

### LDA consistently outperforms LR across the board
### Conclusion: LDA with SMOTE resampling would be the preferred model.
