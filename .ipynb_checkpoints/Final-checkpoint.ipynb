```python
# undersampling the data - splitting into 1 and 0, undersampling not bankrupt, concatenating

by_class = df.groupby('Bankrupt')
bankrupt = by_class.get_group(1)
not_bankrupt = by_class.get_group(0)
not_bankrupt_undersampled = resample(not_bankrupt,random_state=7,n_samples=220,replace=True)
df_undersampled = pd.concat([bankrupt,not_bankrupt_undersampled])
df_undersampled
```

```python
# creating the test and training sets

X = df.loc[:, df.columns != "Bankrupt"]
y = df["Bankrupt"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
%%time

# random forest classifier with default parameters
rf_model = RandomForestClassifier(random_state=42)

# Fit the model to the training data
rf_model.fit(X_train, y_train)

# Make predictions on the test set
predictions = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```

```python
%%time

# random forest classifier with non-default parameters
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

# Fit the model to the training data
rf_model.fit(X_train, y_train)

# Make predictions on the test set
predictions = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```

```python
#displaying the decision rules

def print_decision_rules(rf):

    for tree_idx, est in enumerate(rf.estimators_):
        tree = est.tree_
        assert tree.value.shape[1] == 1 # no support for multi-output

        print('TREE: {}'.format(tree_idx))

        iterator = enumerate(zip(tree.children_left, tree.children_right, tree.feature, tree.threshold, tree.value))
        for node_idx, data in iterator:
            left, right, feature, th, value = data

            # left: index of left child (if any)
            # right: index of right child (if any)
            # feature: index of the feature to check
            # th: the threshold to compare against
            # value: values associated with classes            

            # for classifier, value is 0 except the index of the class to return
            class_idx = np.argmax(value[0])

            if left == -1 and right == -1:
                print('{} LEAF: return class={}'.format(node_idx, class_idx))
            else:
                print('{} NODE: if feature[{}] < {} then next={} else next={}'.format(node_idx, feature, th, left, right))    


rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

print_decision_rules(rf_model)
```

```python
# Yes, the decision rules make sense for this classifier.
```
