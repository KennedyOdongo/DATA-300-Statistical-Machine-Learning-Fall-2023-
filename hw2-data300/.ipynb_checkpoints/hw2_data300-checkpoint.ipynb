2.	Use the KPSS and ADF test to determine whether the is variable stationary. Report the test statistics, the p value, and your conclusions. (4 points) 

```python
electric_df["DATE"] = pd.to_datetime(electric_df["DATE"]) # Python can now read this column as a date column.
```

```python
# Check for the accuracy of the data type
electric_df.dtypes
```

**Sort the DataFrame by the date column so that the variables are arranged in chronological order**

```python
electric_df = electric_df.sort_values("DATE") # This function is used to sort values, with "auto" set to ascending=True.
```

**This function returns the ADF (Augmented Dickey Fuller) value at the 5% significance level**

```python
def adfuller_test(timeseries):
    result = adfuller(timeseries) 
    # adfuller function returns adf (The test statistic), p-value (MacKinnon’s approximate p-value), 
    # usedlag (The number of lags used), 
    # and nobs (The number of observations used for the ADF regression and calculation of the critical values)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        # This loop iterates through the `result` and `labels` lists simultaneously using the `zip` function, 
        # pairing each value from `result` with its corresponding label.
        print(label+' : '+str(value))
        
    if result[1] <= 0.05:
        # A p-value less than 0.05 indicates that we have enough evidence to reject the null hypothesis, 
        # suggesting that the data is stationary
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data is stationary")
    else:
        print("weak evidence against null hypothesis, indicating it is non-stationary ")
```

```python
# Below loop thorugh all the columns in the data set and check whether each variable is stationary or not -using the function above
for i in electric_df.columns:
    print(i), adfuller_test(electric_df[i])
    print()
```

**To check for stationarity using the KPSS test**

```python
for i in electric_df.columns:
    print (i)
    print()
    print (stattools.kpss(electric_df[i], regression='ct'))
    # stattools.kpss returns kpss_stat (The KPSS test statistic), p_value (The p-value of the test), 
    # lags (The truncation lag parameter), and crit (The critical values at 10%, 5%, 2.5% and 1%)
    # “ct” : The data is stationary around a trend.
    print ()
```

- **The p-value (0.9453) is greater than the significance level of 0.05, suggesting weak evidence against the null hypothesis. Therefore, based on the Augmented Dickey-Fuller (ADF) test, the "DATE" variable is likely non-stationary.** 

- **The p-value (0.1862) is greater than the significance level of 0.05, indicating weak evidence against the null hypothesis. Therefore, based on the ADF test, the "IPG2211A2N" variable is likely non-stationary.**

- **Therefore, we need to apply differencing to our model to convert a non-stationary time series into one that exhibits stationarity and stabilizes the mean of the time series.**

3.	Plot the variable and label your graph appropriately. Plot the variable after differencing it once. Plot the variable after differencing it twice. Describe the three graphs (3 points)

```python
# Remove the date column because we cannot plot it
electric_no_date = electric_df.drop(columns ='DATE')
```

**Plot raw variable**

```python
for i, col in enumerate(electric_no_date.columns):
    # enumerate function is used to iterate over elements of an iterable.
    electric_no_date[col].plot(fig=plt.figure(i))
    plt.title(col)

plt.show()
```

**Plot once differenced variable**

```python
for i, col in enumerate(electric_no_date.columns):
    electric_no_date[col].diff().plot(fig=plt.figure(i))
    plt.title(col)

plt.show() # electric_no_date[col].diff()
```

**Plot twice differenced variable**

```python
for i, col in enumerate(electric_no_date.columns):
    electric_no_date[col].diff().diff().plot(fig=plt.figure(i))
    plt.title(col)

plt.show() # electric_no_date[col].diff().diff()
```

**As we can observe, after differencing the variable, the data becomes more stationary. The more we difference the variable, the more stationary it becomes. However, the once-differenced variable and twice-differenced variable are not significantly different from each other.**

4.	Plot the PACF and ACF graphs for the variables. Determine the ARMA order (4 points)

```python
# This function is designed to create and display both the Autocorrelation Function (ACF) plot
# and the Partial Autocorrelation Function (PACF) plot for a given time series dataset.
def acf_pacf (series)-> plt.plot:
    return plot_acf(series), plot_pacf(series)
# plot_acf function is from statsmodels, creating the ACF plot. 
# plot_pacf function is from statsmodels, showing the autocorrelation of the time series at different lags (time lags).
```

```python
# Plot the ACF and PACF plots for all the series in the data
for i in electric_df.columns:
    print(i)
    acf_pacf(electric_df[i])
    print()
```

```python
def arimamodel(timeseries):
    # auto_arima function from the pmdarima library to automatically determine the best ARIMA model for the given timeseries
    automodel = pm.auto_arima(timeseries, 
                              start_p=1, # Specify the range for the number of autoregressive (p) 
                              start_q=1, # and moving average (q) terms to consider during the model selection process
                              test="adf", # Specify the method to use for the differencing test (adf)
                              seasonal=True, # Consider seasonal components in the time series
                              trace=True) # Enables tracing and provides detailed information about the model selection process
    return automodel
```

```python
columns = ['IPG2211A2N']
```

```python
for i in columns:
    print(i)
    arimamodel(electric_df[i])
    print()
```

6.	Fit your model based on the training set that you determine (70%, 80% or 90% of the data). (1 point)

```python
model = sm.tsa.arima.ARIMA(electric_df['IPG2211A2N'].head(int(0.8*len(electric_df))), order=(4, 1, 2))
results = model.fit()
```

```python
results.summary()
```

7.	Forecast out of sample for the length of your test set (1 point)

**int(0.8*len(electric_df)) is the training set, which are the first 80% of data points for the IPG2211A2N. You want the test set to start right after it.**

```python
electric_df['IPG2211A2N_forecast']=results.predict(start = int(0.8*len(electric_df))+1, end=len(electric_df) , dynamic=True)
```
